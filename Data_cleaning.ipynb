```python
import pandas as pd
print(pd.__version__)
print(pd.__file__)
```

```python
df = pd.read_excel("C:/Users/HP/Desktop/Data_Analytics.xlsx")
print("Data Imported Successfuly")
```

```python
# Quick Inspection of the data
print(df.head())
print(df.tail())
df.info()
df.isnull().sum()
df.isnull().sum().sum()
df.describe()
```

```python
# Handling Missing Value (Phase 1)
# Fill the missing Coupon codes with the string NoCode

df['CouponCode'] = df['CouponCode'].fillna("NoCode")
```

```python
# Removing Duplicates (Phase 2)
# Order IDs must be unique

df.duplicated().sum()

duplicate_count = df['OrderID'].duplicated().sum()
print(duplicate_count)

# There are no duplicate rows and the OrderIDs are truly unique
```

```python
# Speaking the same language (Phase 3)

# Convert the date to ISO 8601 format
df["Date"] = pd.to_datetime(df["Date"]).dt.strftime("%Y-%m-%d")
```

```python
# Finding all text columns
text_colums = df.select_dtypes(include=["object"]).columns
print("Text columns in my data: ")
print(list(text_colums))
```

```python
# Finding col with leading or trailing spaces 
for col in text_colums:
    has_spaces = df[col].astype(str).str.strip().ne(df[col].astype(str)).any()
    if has_spaces:
        print(f"'{col}' has extra spaces that needs trimming")

# Trimming the columns
for col in text_colums:
    df[col] = df[col].str.strip()        
```

```python
# Check for mixed case issues
for col in text_colums :
    unique_val = df[col].dropna().unique()
    lower_val = [str(x).lower() for x in unique_val]
    if len(set(lower_val)) < len(unique_val):
        print(f"\n' {col}' has case inconsistencies: ")
        for val in unique_val:
            print(f" -{val}")
```
